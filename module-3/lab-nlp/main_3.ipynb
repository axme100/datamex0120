# Create and test models

```python
import pickle
import nltk
```

```python
# Import the corpus as features
pickle_in = open("feature_set","rb")
features = pickle.load(pickle_in)
```

```python
# Check to make sure the features are a list
type(features)
```

```python
# Check that we have all the features
len(features)
```

```python
# Check to make sure the we have the correct structure of features
# To feed into the algorithm
features[0]
```

```python
# Manually split up the data into a training and a testing data set
# 20% for the training set
training_set = features[:1280000]
testing_set = features[1280000:]
```

```python
# Train the classifier
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
# Show the accuracy of the Naive Bayes Algorithm agasint the testing set
print("Naive Bayes Accuracy: ")
print(nltk.classify.accuracy(classifier, testing_set))
```
