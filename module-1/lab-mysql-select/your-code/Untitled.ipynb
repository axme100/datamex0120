# Solution to Challenge 1

```python
query = ''' SELECT TWO.id_author as 'AUTHOR ID', TWO.lastname as 'LAST NAME', TWO.firstname as 'FIRSTNAME', publishers.pub_name as 'PUBLISHER'
                FROM (SELECT ONE.id_author, ONE.id_title, ONE.firstname, ONE.lastname, titles.title as 'book_title', titles.pub_id as 'id_pub'
                FROM (SELECT titleauthor.au_id as 'id_author', title_id as 'id_title', au_lname as 'lastname', au_fname as 'firstname'
                FROM titleauthor
                LEFT JOIN authors
                ON titleauthor.au_id = authors.au_id) AS ONE
            LEFT JOIN titles
            ON ONE.id_title = titles.title_id) AS TWO
            LEFT JOIN publishers
            ON TWO.id_pub = publishers.pub_id'''
```

```python
pd.read_sql(query, engine)
```

# Solution to Challenge 2

```python
query = ''' SELECT AUTHOR_ID, LAST_NAME, FIRSTNAME, PUBLISHER, COUNT(*) as 'TITLE_COUNT'
            FROM (SELECT TWO.id_author as 'AUTHOR_ID', TWO.lastname as 'LAST_NAME', TWO.firstname as 'FIRSTNAME', publishers.pub_name as 'PUBLISHER'
                FROM (SELECT ONE.id_author, ONE.id_title, ONE.firstname, ONE.lastname, titles.title as 'book_title', titles.pub_id as 'id_pub'
                FROM (SELECT titleauthor.au_id as 'id_author', title_id as 'id_title', au_lname as 'lastname', au_fname as 'firstname'
                FROM titleauthor
                LEFT JOIN authors
                ON titleauthor.au_id = authors.au_id) AS ONE
            LEFT JOIN titles
            ON ONE.id_title = titles.title_id) AS TWO
            LEFT JOIN publishers
            ON TWO.id_pub = publishers.pub_id) AS THREE
            GROUP BY AUTHOR_ID, PUBLISHER
            ORDER BY TITLE_COUNT DESC;'''
```

```python
pd.read_sql(query, engine)
```

```python
# Join titleauthor and sales
query = '''SELECT sales.qty, titleauthor.au_id
           FROM sales
           LEFT JOIN titleauthor
           ON sales.title_id = titleauthor.title_id
'''

pd.read_sql(query, engine)
```

```python
# Join result to author
# Join titleauthor and sales
query = '''
           SELECT id_author, authors.au_lname as 'lastname', authors.au_fname as 'firstname', qty
              FROM (SELECT sales.qty, titleauthor.au_id as 'id_author'
              FROM sales
              LEFT JOIN titleauthor
              ON sales.title_id = titleauthor.title_id) AS ONE
           LEFT JOIN authors
           ON id_author = authors.au_id
'''

pd.read_sql(query, engine)
```

```python
# Sum the QTY
# Join result to author
# Join titleauthor and sales
query = '''
           SELECT id_author, authors.au_lname as 'lastname', authors.au_fname as 'firstname', sum(qty) as 'qty'
              FROM (SELECT sales.qty, titleauthor.au_id as 'id_author'
              FROM sales
              LEFT JOIN titleauthor
              ON sales.title_id = titleauthor.title_id) AS ONE
           LEFT JOIN authors
           ON id_author = authors.au_id
           GROUP BY id_author
           ORDER BY qty DESC;
'''

pd.read_sql(query, engine)
```

```python
# Sum the QTY
# Join result to author
# Join titleauthor and sales
query = '''
           SELECT id_author, authors.au_lname as 'lastname', authors.au_fname as 'firstname', IFNULL(SUM(qtysales), 0) AS 'total'
              FROM (SELECT sales.qty as 'qtysales', titleauthor.au_id as 'id_author'
              FROM titleauthor
              LEFT JOIN sales
              ON titleauthor.title_id = sales.title_id) AS ONE
           LEFT JOIN authors
           ON id_author = authors.au_id
           GROUP BY id_author;
        
'''

pd.read_sql(query, engine)
```
