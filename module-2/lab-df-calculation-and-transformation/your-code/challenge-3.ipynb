## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

```python
Aquí es para mostrar que hay clientes en diferentes países

# Agrupar por país
countryCustomer = orders.groupby(['Country', 'CustomerID'])[['amount_spent']].sum()

# Hay clientes que están en varios lugares.
len(countryCustomer['amount_spent'].unique())
```

```python
# your code here
customerSpent = orders.groupby(['CustomerID'])[['amount_spent']].sum()
customerSpent.head()
```

```python
sevFiveQuan = customerSpent.quantile(.75)[0]
sevFiveQuan
```

```python
nineFiveQuan = customerSpent.quantile(.95)[0]
nineFiveQuan
```

```python
# At first, I wanted to do this with a lambda:
# https://stackoverflow.com/a/31294135/5420796
def assignCustomerLevel(amount_spent):
    if amount_spent > nineFiveQuan:
        return "vip"
    elif amount_spent > sevFiveQuan:
        return "preferred"
    else:
        return "none"
```

```python
customerSpent['customerLevel'] = customerSpent['amount_spent'].apply(assignCustomerLevel)
```

```python
customerSpentCountry = pd.merge(customerSpent, orders[['CustomerID', 'Country']], on='CustomerID')
```

```python
vips = customerSpentCountry[customerSpentCountry['customerLevel'] == "vip"]
```

```python
preferred = customerSpentCountry[customerSpentCountry['customerLevel'] == "preferred"]
```

```python
preferredORVIP = customerSpentCountry[(customerSpentCountry['customerLevel'] == "preferred") | (customerSpentCountry['customerLevel'] == "vip")]
```
